```python
# For data
import pandas as pd
from pandas import Series,DataFrame
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from __future__ import division

# Use to grab data from the web(HTTP capabilities)
import requests

# We'll also use StringIO to work with the csv file, the DataFrame will require a .read() method
from StringIO import StringIO
```

```python
url = "http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv"

# Use requests to get the information in text form
source = requests.get(url).text

# Use StringIO to avoid an IO error with pandas
poll_data = StringIO(source) 
```

```python
poll_df = pd.read_csv(poll_data)
poll_df.info()
```

```python
poll_df = poll_df.drop('Question Iteration', 1)
```

```python
poll_df.info()
```

```python
poll_df.head()
```

```python
# Factorplot the affiliation
sns.countplot('Affiliation',data=poll_df)
```

```python
# Factorplot the affiliation by Population
sns.countplot('Affiliation',data=poll_df,hue='Population')
```

```python
#Did the poll results favor Romney or Obama?

# First we'll get the average
avg = pd.DataFrame(poll_df.mean())
avg.head()
```

```python
# Drop Number of Observations and Others
avg.drop('Number of Observations',axis=0,inplace=True)
avg.drop('Other', axis=0,inplace=True)

# After that let's get the error
std = pd.DataFrame(poll_df.std())
std.drop('Number of Observations',axis=0,inplace=True)
std.drop('Other', axis=0,inplace=True)

# now plot using pandas built-in plot, with kind='bar' and yerr='std'
avg.plot(yerr=std,kind='bar',legend=False)
```

```python
# Concatenate the Average and Std DataFrames
poll_avg = pd.concat([avg,std],axis=1)
poll_avg.columns = ['Average','STD']
poll_avg
```

```python
# How do undecided voters effect the poll? Can we account for the undecided voters?
# If we assume we split the undecided evenly between the two candidates the observed difference should be an unbiased estimate of the final difference.

# Quick plot of sentiment in the polls versus time.
poll_df.plot(x='End Date',y=['Obama','Romney','Undecided'],marker='o',linestyle='')
```

```python
# How did voter sentiment change over time? 

# For timestamps
from datetime import datetime
```

```python
# Now we'll define a new column in our poll_df DataFrame to take into account the difference between Romney and Obama in the polls. The Difference column is Obama minus Romney in percentage, thus a positive difference indicates a leaning towards Obama in the polls.

poll_df['Difference'] = (poll_df.Obama - poll_df.Romney)/100
poll_df.head()
```

```python
# Visualisation of the this sentiment in difference changes over time (poll_df): 

# First: groupby to group the polls by their start data and then sorting it by that Start Date.
poll_df = poll_df.groupby(['Start Date'],as_index=False).mean()
poll_df.head()
```

```python
# Second: # Plotting the difference in polls between Obama and Romney

fig = poll_df.plot('Start Date','Difference',figsize=(16,6),marker='o',linestyle='-')
```

```python
# Can we see an effect in the polls from the debates? (The debate dates were Oct 3rd, Oct 11, and Oct 22nd.)

# In order to find where to set the x limits for the figure we need to find out where the index for the month of October in 2012 is. Here's a simple for loop to find that row. Note, the string format of the date makes this difficult to do without using a lambda expression or a map.

# Set row count and xlimit list
row_in = 0
xlimit = []

# Cycle through dates until 2012-10 is found, then print row index
for date in poll_df['Start Date']:
    if date[0:7] == '2012-10':
        xlimit.append(row_in)
        row_in +=1
    else:
        row_in += 1
        
print min(xlimit)
print max(xlimit)
```
